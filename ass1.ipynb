```python
import numpy as np #Here we use the module numpy to help us generate random variables and denote it as "np".
import matplotlib.pyplot as plt #Matplotlib is a 2D drawing library for Python
from scipy.stats import norm

#quesition 1
#question (a)
fig, axes = plt.subplots(2,2) # create blank subplots
n = 100
data = np.random.exponential(2,n)
axes[0,0].hist(data, density=True,bins=40)
lambd = 2
x = np.arange(0,10,0.1)
y = lambd*np.exp(-lambd*x)
axes[0,0].plot(x,y)

#question (b)
mean = np.mean(data)
var = np.var(data)
T=np.sqrt(n)*(mean-2)/(np.sqrt(var))
print('mean is {}, variance is {},T statistics is{}'.format(mean, var, T)) 

#question (c)
data1 = []
for i in range(1,1000):
    data = np.random.exponential(2,n)
    mean = np.mean(data)
    var = np.var(data)
    T=np.sqrt(n)*(mean-2)/(np.sqrt(var))
    data1.append(T)
axes[0,1].hist(data1, density=True,bins=100)
x = np.arange(-4,4,0.1)
y = np.exp(-np.square(x)/2)/np.sqrt(2*np.pi)
axes[0,1].plot(x,y)

#question (d)
data2 = []
for i in data1:
    data2.append(1-norm.cdf(i))
axes[1,0].hist(data2, density=True,bins=100)
axes[1,0].plot([0,1],[1,1])
```

```python
#question 2
#question (a)
x = np.random.uniform(low = 0, high = 1.0, size = 100) # generate datas following uniform disribution
eps = np.random.normal(size = 100) 
y = np.sin(2 * np.pi * x) +np.cos(2 * np.pi * x)+ eps #calculate the function values
fig, axes = plt.subplots(2,2) # create blank subplots
axes[0,0].scatter(x, y) # present the first scatter plot with Q1

#question (b)
xs = np.linspace(start = 0, stop = 1.0, num = 20) 
h1=0.05
h2=0.15
h3=0.25
ret1 = np.zeros(20)
ret2 = np.zeros(20)
ret3 = np.zeros(20)
for i in range(20):
    z=xs[i]
    ret1[i] = np.sum((np.absolute(x - z) <= h1) * 1.0 * y) / (2.0 * 100 * h1) #calculate the estimated function values
    ret2[i] = np.sum((np.absolute(x - z) <= h2) * 1.0 * y) / (2.0 * 100 * h2) #calculate the estimated function values
    ret3[i] = np.sum((np.absolute(x - z) <= h3) * 1.0 * y) / (2.0 * 100 * h3) #calculate the estimated function values
axes[0,1].plot(xs, ret1, label = 'h = {}'.format(h1)) 
axes[0,1].plot(xs, ret2, label = 'h = {}'.format(h2)) 
axes[0,1].plot(xs, ret3, label = 'h = {}'.format(h3)) 
axes[0,1].legend()

#question (c)
ret4 = np.zeros(20)
ret5 = np.zeros(20)
ret6 = np.zeros(20)
for i in range(20):
    z=xs[i]
    ret4[i] = np.sum(np.exp(-np.square((x-z)/h1)/2)/np.sqrt(2*np.pi) * y) / (2.0 * 100 * h1) #calculate the estimated function values
    ret5[i] = np.sum(np.exp(-np.square((x-z)/h2)/2)/np.sqrt(2*np.pi) * y) / (2.0 * 100 * h2) #calculate the estimated function values
    ret6[i] = np.sum(np.exp(-np.square((x-z)/h3)/2)/np.sqrt(2*np.pi) * y) / (2.0 * 100 * h3) #calculate the estimated function values
axes[1,0].plot(xs, ret4, label = 'h = {}'.format(h1)) 
axes[1,0].plot(xs, ret5, label = 'h = {}'.format(h2)) 
axes[1,0].plot(xs, ret6, label = 'h = {}'.format(h3)) 
axes[1,0].legend()
```
