```python
#   Q1
import numpy as np
import matplotlib.pyplot as plt
beta0 = 1
beta1 = 2
n = 20
residual = np.random.normal(0,0.1,n)
X = np.random.uniform(0,1,n)
Y = beta0 + beta1*X + residual
```

```python
#    1)
fig, axes = plt.subplots(1,2,constrained_layout = True,figsize = (8,3))
axes[0].scatter(X,Y,s=15)
#    2)
X1 = np.zeros((n,2))
for i in range(n):
    X1[i] = [1, X[i]]
beta_estimater = np.dot(np.linalg.inv(np.dot(X1.T,X1)/n),(np.dot(X1.T,Y)/n))
print(beta_estimater)
x = np.linspace(0,1)
y = beta_estimater[0] + beta_estimater[1]*x
axes[1].scatter(X,Y,s=15)
axes[1].plot(x,y,linewidth = 1)
```
