```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x = np.linspace(start = 0.01, stop = 1.0, num = 100)
n1 = 1
n2 = 5
n3 = 10
n4 = 30
lamda = 1/10
y1 = np.zeros(100)
y2 = np.zeros(100)
y3 = np.zeros(100)
y4 = np.zeros(100)

for i in range(100):
    z = x[i]
    y1[i] = (z**n1)*math.exp(-10*n1*z)-(lamda**n1)*math.exp(-10*n1*lamda)
    y2[i] = (z**n2)*math.exp(-10*n2*z)-(lamda**n2)*math.exp(-10*n2*lamda)
    y3[i] = (z**n3)*math.exp(-10*n3*z)-(lamda**n3)*math.exp(-10*n3*lamda)
    y4[i] = (z**n4)*math.exp(-10*n4*z)-(lamda**n4)*math.exp(-10*n4*lamda)

fig, axes = plt.subplots(4,1,constrained_layout = True,figsize = (6,8)) # create blank subplots
axes[0].set_title('n = 1')
axes[1].set_title('n = 5')
axes[2].set_title('n = 10')
axes[3].set_title('n = 30')
axes[0].plot(x, y1, color = 'steelblue') 
axes[1].plot(x, y2, color = 'green') 
axes[2].plot(x, y3, color = 'darkorange')
axes[3].plot(x, y4, color = 'hotpink') 
```

```python
z = 1.96
CI = np.array([
[lamda-z*lamda/math.sqrt(n1), lamda+z*lamda/math.sqrt(n1)],
[lamda-z*lamda/math.sqrt(n2), lamda+z*lamda/math.sqrt(n2)],
[lamda-z*lamda/math.sqrt(n3), lamda+z*lamda/math.sqrt(n3)],
[lamda-z*lamda/math.sqrt(n4), lamda+z*lamda/math.sqrt(n4)]])
print(CI)
```

```python

sample_size = 10000
lamb0 = 1
for i in range(100):
    y = np.random.exponential(lamb0, sample_size)
    mean = np.mean(y)
    confidence_interval = [1/mean-z/(mean*math.sqrt(sample_size)), 1/mean+z/(mean*math.sqrt(sample_size))]
    if confidence_interval[0] <= lamb0 and confidence_interval[1] >= lamb0:
        color = 'grey'
    else:
        color = 'red'
    plt.hlines(i, confidence_interval[0], confidence_interval[1], color = color)
plt.vlines(lamb0, 0, 100, color='green')
plt.xlabel("lambda")
plt.ylabel("Samples")
```
